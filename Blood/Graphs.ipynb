```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df_40 = pd.read_csv('cali_40.csv')
df_80 = pd.read_csv('cali_80.csv')
df_120 = pd.read_csv('cali_120.csv')
df_160 = pd.read_csv('cali_160.csv')
df_rest = pd.read_csv('rest_pressure.csv')
df_hard = pd.read_csv('hard_pressure.csv')
filtered_df_hard = df_hard[(df_hard['Время[с]'] >= 8) & (df_hard['Время[с]'] <= 18)]
filtered_df_rest = df_rest[(df_rest['Время[с]'] >= 18) & (df_rest['Время[с]'] <= 28)]
filtered_df_40 = df_40[(df_40['Время[с]'] >= 2)]
filtered_df_160 = df_160[(df_160['Время[с]'] >= 6)]

df_40.head()
plt.figure(figsize=(10, 6))
plt.plot(df_40['Время[с]'], df_40['Напряжение[В]'])
plt.title('Калибровка при 40 мм рт. ст.')
plt.xlabel('Время[c]')
plt.ylabel('Напряжение[B]')
plt.grid(True)
plt.ylim(0.0, 0.9)
plt.savefig('/prog/labs/Blood/cali_40.png')
plt.show()
mean_value_40 = filtered_df_40['Напряжение[В]'].mean()
print(f"Среднее значение напряжения: {mean_value_40} В")

plt.figure(figsize=(10, 6))
plt.plot(df_80['Время[с]'], df_80['Напряжение[В]'])
plt.title('Калибровка при 80 мм рт. ст.')
plt.xlabel('Время[c]')
plt.ylabel('Напряжение[B]')
plt.grid(True)
plt.ylim(0.4, 1.3)
plt.savefig('cali_80.png')
plt.show()
mean_value_80 = df_80['Напряжение[В]'].mean()
print(f"Среднее значение напряжения: {mean_value_80} В")

plt.figure(figsize=(10, 6))
plt.plot(df_120['Время[с]'], df_120['Напряжение[В]'])
plt.title('Калибровка при 120 мм рт. ст.')
plt.xlabel('Время[c]')
plt.ylabel('Напряжение[B]')
plt.grid(True)
plt.ylim(0.8, 1.7)
plt.savefig('cali_120.png')
plt.show()
mean_value_120 = df_120['Напряжение[В]'].mean()
print(f"Среднее значение напряжения: {mean_value_120} В")

plt.figure(figsize=(10, 6))
plt.plot(df_160['Время[с]'], df_160['Напряжение[В]'])
plt.title('Калибровка при 160 мм рт. ст.')
plt.xlabel('Время[c]')
plt.ylabel('Напряжение[B]')
plt.grid(True)
plt.ylim(1.2, 2.1)
plt.savefig('cali_160.png')
plt.show()
mean_value_160 = filtered_df_160['Напряжение[В]'].mean()
print(f"Среднее значение напряжения: {mean_value_160} В")

plt.figure(figsize=(10, 6))
plt.plot(df_rest['Время[с]'], df_rest['Напряжение[В]'])
plt.title('Эксперимент в покое')
plt.xlabel('Время[c]')
plt.ylabel('Напряжение[B]')
plt.grid(True)
#plt.ylim(1.2, 2.1)
plt.savefig('rest_voltage.png')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(df_hard['Время[с]'], df_hard['Напряжение[В]'])
plt.title('Эксперимент в возбужденном состоянии')
plt.xlabel('Время[c]')
plt.ylabel('Напряжение[B]')
plt.grid(True)
#plt.ylim(1.2, 2.1)
plt.savefig('hard_voltage.png')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(filtered_df_rest['Время[с]'], filtered_df_rest['Напряжение[В]'])
plt.title('Эксперимент в покое (прямой участок)')
plt.xlabel('Время[c]')
plt.ylabel('Напряжение[B]')
plt.grid(True)
#plt.ylim(1.2, 2.1)
plt.savefig('filt_rest_voltage.png')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(filtered_df_hard['Время[с]'], filtered_df_hard['Напряжение[В]'])
plt.title('Эксперимент в возбужденном состоянии (прямой участок)')
plt.xlabel('Время[c]')
plt.ylabel('Напряжение[B]')
plt.grid(True)
#plt.ylim(1.2, 2.1)
plt.savefig('filt_hard_voltage.png')
plt.show()




x1 = [mean_value_40, mean_value_80, mean_value_120, mean_value_160]
y = [40, 80, 120, 160]
k, b = np.polyfit(x1, y, 1)
plt.figure(figsize=(10, 6))
plt.scatter(x1, y, alpha=0.7, label='Экспериментальные данные')
x_line = np.linspace(min(x1), max(x1), 100)
y_line = k * x_line + b
plt.plot(x_line, y_line, color='red', linewidth=2, label=f'Калибровочный коэффициент: y = {k:.2f}x  {b:.2f}')
plt.title('Калибровка')
plt.xlabel('Среднее напряжение, В')
plt.ylabel('Давление, мм рт. ст.')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('cali_k.png')
plt.show()


df_hard['Напряжение[В]'] = df_hard['Напряжение[В]']*k+b
df_rest['Напряжение[В]'] = df_rest['Напряжение[В]']*k+b
filtered_df_rest['Напряжение[В]'] = filtered_df_rest['Напряжение[В]']*k+b
filtered_df_hard['Напряжение[В]'] = filtered_df_hard['Напряжение[В]']*k+b

plt.figure(figsize=(10, 6))
plt.plot(df_rest['Время[с]'], df_rest['Напряжение[В]'])
plt.scatter(5, 101, color='red', zorder=3)
plt.text(
    6, 102,             # координаты подписи рядом
    "Систола", 
    va='center',
    ha='left'
)
plt.scatter(40.9, 56, color='orange', zorder=3)
plt.text(
    34, 54,             # координаты подписи рядом
    "Диастола", 
    va='center',
    ha='left'
)
plt.title('Эксперимент в покое')
plt.xlabel('Время[c]')
plt.ylabel('Артериальное давление[мм рт. ст.]')
plt.grid(True)
#plt.ylim(1.2, 2.1)
plt.savefig('rest_pressure.png')
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(df_hard['Время[с]'], df_hard['Напряжение[В]'])
plt.title('Эксперимент в возбужденном состоянии')
plt.xlabel('Время[c]')
plt.ylabel('Артериальное давление[мм рт. ст.')
plt.scatter(1.2, 109.5, color='red', zorder=3)
plt.text(
    2.3, 111,             # координаты подписи рядом
    "Систола", 
    va='center',
    ha='left'
)
plt.scatter(21, 59, color='orange', zorder=3)
plt.text(
    14.5, 56,             # координаты подписи рядом
    "Диастола", 
    va='center',
    ha='left'
)
plt.grid(True)
#plt.ylim(1.2, 2.1)
plt.savefig('hard_pressure.png')
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(filtered_df_rest['Время[с]'], filtered_df_rest['Напряжение[В]'])
plt.title('Эксперимент в покое (прямой участок)')
plt.xlabel('Время[c]')
plt.ylabel('Артериальное давление[мм рт. ст.]')
plt.grid(True)
#plt.ylim(1.2, 2.1)
plt.savefig('filt_rest_pressure.png')
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(filtered_df_hard['Время[с]'], filtered_df_hard['Напряжение[В]'])
plt.title('Эксперимент в возбужденном состоянии (прямой участок)')
plt.xlabel('Время[c]')
plt.ylabel('Артериальное давление[мм рт. ст.')
plt.grid(True)
#plt.ylim(1.2, 2.1)
plt.savefig('filt_hard_pressure.png')
plt.show()


k1, b1 = np.polyfit(filtered_df_rest['Время[с]'], filtered_df_rest['Напряжение[В]'], 1)
plt.figure(figsize=(10, 6))
filt_df_rest = filtered_df_rest['Напряжение[В]'] - filtered_df_rest['Время[с]']*k1-b1
plt.plot(filtered_df_rest['Время[с]'], filt_df_rest)
plt.title('Пульс в спокойном состоянии')
plt.xlabel('Время, с')
plt.ylabel('Изменения давления, мм рт. ст.')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(-10, 10)
plt.savefig('pulse_rest.png')
plt.show()


k2, b2 = np.polyfit(filtered_df_hard['Время[с]'], filtered_df_hard['Напряжение[В]'], 1)
plt.figure(figsize=(10, 6))
filt_df_hard = filtered_df_hard['Напряжение[В]'] - filtered_df_hard['Время[с]']*k2-b2
plt.plot(filtered_df_hard['Время[с]'], filt_df_hard)
plt.title('Пульс в возбужденном состоянии')
plt.xlabel('Время, с')
plt.ylabel('Изменения давления, мм рт. ст.')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(-10, 10)
plt.savefig('pulse_hard.png')
plt.show()
```
